Observamos que entre 2007 e 1997 a população aumentou 21.463.928. Sendo assim, consideremos uma situação hipotética onde a cada 10 anos a população cresça 21.463.928. Dessa maneira, podemos descobrir o momento em que o Brasil alcançará 500.000.000 de habitantes.



> Utilizaremos os laços de programação para realizar essa atividade:



```python
P2007 = 190010647
TP = 0

while P2007 < 500000000:
  P2007 = P2007 + 21463928
  for i in [10]:
    print("Após", i, "anos teremos", P2007)
    TP = TP + 1
print("O tempo total foi de", TP*10,"anos")
```

A partir da análise acima, compreende-se que nessa situação hipotética o Brasil levaria cerca de 150 anos para alcançar uma população de 500.000.000.

Portanto, levando toda a análise realizada em consideração, podemos extrair as seguintes conclusões:


*   A população mundial aumentou entre 1952 e 2007.
*   De maneira geral, o PIB per capita dos países aumentou.
*   A expectativa de vida dos Continentes Americano e Asiático aumentou.
*   O PIB per capita não possui uma FORTE influência sobre a expectativa de vida devido à desigualdade socioeconômica.
*   Entre 1952 e 2007 a população brasileira cresceu de forma quase linear.
